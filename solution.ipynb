```python
from __future__ import absolute_import

import random

import matplotlib.pyplot as plt
import numpy as np
from numpy import *
import scipy.io
import scipy.misc


def plot_data(samples, centroids, clusters=None):
    """
    Plot samples and color it according to cluster centroid.
    :param samples: samples that need to be plotted.
    :param centroids: cluster centroids.
    :param clusters: list of clusters corresponding to each sample.
    """

    colors = ['blue', 'green', 'gold']
    assert centroids is not None

    if clusters is not None:
        sub_samples = []
        for cluster_id in range(centroids[0].shape[0]):
            sub_samples.append(np.array([samples[i] for i in range(samples.shape[0]) if clusters[i] == cluster_id]))
    else:
        sub_samples = [samples]

    plt.figure(figsize=(7, 5))

    for clustered_samples in sub_samples:
        cluster_id = sub_samples.index(clustered_samples)
        plt.plot(clustered_samples[:, 0], clustered_samples[:, 1], 'o', color=colors[cluster_id], alpha=0.75,
                 label='Data Points: Cluster %d' % cluster_id)

    plt.xlabel('x1', fontsize=14)
    plt.ylabel('x2', fontsize=14)
    plt.title('Plot of X Points', fontsize=16)
    plt.grid(True)

    # Drawing a history of centroid movement
    tempx, tempy = [], []
    for mycentroid in centroids:
        tempx.append(mycentroid[:, 0])
        tempy.append(mycentroid[:, 1])

    for cluster_id in range(len(tempx[0])):
        plt.plot(tempx, tempy, 'rx--', markersize=8)

    plt.legend(loc=4, framealpha=0.5)
    plt.show(block=True)


def get_centroids(samples, clusters):
    """
    Find the centroid given the samples and their cluster.

    :param samples: samples.
    :param clusters: list of clusters corresponding to each sample.
    :return: an array of centroids.
    """
    # print(clusters)
    # print(samples)
    i=0
    cluster0=0
    cluster1=0
    cluster2=0
    sum_cluster0_x1=0
    sum_cluster0_x2=0
    sum_cluster1_x1=0
    sum_cluster1_x2=0
    sum_cluster2_x1=0
    sum_cluster2_x2=0
    for i in range(len(clusters)):                              #calculate sum for each cluster, and number of samples for each cluster
        if(clusters[i] == 0):
            cluster0=cluster0+1
            sum_cluster0_x1=sum_cluster0_x1+samples[i, 0]
            sum_cluster0_x2=sum_cluster0_x2+samples[i, 1]
        elif(clusters[i] == 1):
            cluster1=cluster1+1
            sum_cluster1_x1=sum_cluster1_x1+samples[i, 0]
            sum_cluster1_x2=sum_cluster1_x2+samples[i, 1]
        else:
            cluster2=cluster2+1
            sum_cluster2_x1=sum_cluster2_x1+samples[i, 0]
            sum_cluster2_x2=sum_cluster2_x2+samples[i, 1]
    # print(sum_cluster2_x1)
    # print(cluster2)                                                   #calculate centroid
    centroids=[[sum_cluster0_x1/cluster0,sum_cluster0_x2/cluster0],[sum_cluster1_x1/cluster1,sum_cluster1_x2/cluster1],[sum_cluster2_x1/cluster2,sum_cluster2_x2/cluster2]]
    centroids=np.array(centroids)
    print(centroids)
    return centroids

def find_closest_centroids(samples, centroids):
    """
    Find the closest centroid for all samples.

    :param samples: samples.
    :param centroids: an array of centroids.
    :return: a list of cluster_id assignment.
    """   
    # print(len(centroids))
    # print(np.shape(samples))
    # print(np.shape(centroids))
    # print(type(samples))
    # print(type(centroids))
    i=0
    cluster_list=[]
    for i in range (len(samples)):
        j=0
        distance_list=[]
        for j in range (len(centroids)):                #clculate the distance to each centroid
            X1=samples[i, 0]-centroids[j, 0]
            X2=samples[i, 1]-centroids[j, 1]
            distance=(X1**2+X2**2)**0.5
            # print(type(X1))
            distance_list.append(distance)
        index=[k for k,x in enumerate(distance_list) if x==min(distance_list)]      #find cluster that has the minimum distance 
        cluster_list=cluster_list+index 
    cluster_array=np.array(cluster_list)
    print(cluster_array)
    return cluster_array


def run_k_means(samples, initial_centroids, n_iter):
    """
    Run K-means algorithm. The number of clusters 'K' is defined by the size of initial_centroids
    :param samples: samples.
    :param initial_centroids: a list of initial centroids.
    :param n_iter: number of iterations.
    :return: a pair of cluster assignment and history of centroids.
    """
    centroid_history = []
    current_centroids = initial_centroids
    clusters = []
    for iteration in range(n_iter):
        centroid_history.append(current_centroids)
        print("Iteration %d, Finding centroids for all samples..." % iteration)
        clusters = find_closest_centroids(samples, current_centroids)
        print("Recompute centroids...")
        current_centroids = get_centroids(samples, clusters)

    return clusters, centroid_history


def choose_random_centroids(samples, K):
    """
    Randomly choose K centroids from samples.
    :param samples: samples.
    :param K: K as in K-means. Number of clusters.
    :return: an array of centroids.
    """ 
    np.random.seed(7)
    np.random.shuffle(samples)          #shuffle the data randomly
    centroids = samples[0:K]            #choose the first K element
    return centroids
    

def main():
    datafile = 'C:\\Users\\dell\\Desktop\\cs506\\hw1\\homework-1-Kotoric1\\kmeans-data.mat'
    mat = scipy.io.loadmat(datafile)
    samples = mat['X']
    # samples contain 300 pts, each has two coordinates

    # Choose the initial centroids
    initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])
    plot_data(samples, [initial_centroids])
    clusters = find_closest_centroids(samples, initial_centroids)

    # you should see the output [0, 2, 1] corresponding to the
    # centroid assignments for the first 3 examples.
    print(np.array(clusters[:3]).flatten())
    plot_data(samples, [initial_centroids], clusters)
    clusters, centroid_history = run_k_means(samples, initial_centroids, n_iter=10)
    plot_data(samples, centroid_history, clusters)

    # Let's choose random initial centroids and see the resulting
    # centroid progression plot.. perhaps three times in a row
    # for x in range(3):
    #     clusters, centroid_history = run_k_means(samples, choose_random_centroids(samples, K=3), n_iter=10)
    #     plot_data(samples, centroid_history, clusters)


if __name__ == '__main__':
    random.seed(7)
    main()




```

### 2. Working with the Algorithms

```python
from typing import List, Dict, Tuple
# feel free to add more functions such as discard missing examples
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture
def read_dataset(dataset_path: str):
    """
    read in NYC dataset and return a dataset type of your choice
    :param dataset_path: the string path to the dataset file
    :return: dataset
    """
    df=pd.read_csv(dataset_path,usecols=[6,7,9])
    data=df.values
    nan_value=[]
    new_data=[]
    for i in range(len(data)):                      #read the data and delete the nan values
        for j in range(3):
            if (data[i][j]==0):
                nan_value.append(i)
    for k in range (len(data)):
        if k not in nan_value:
            new_data.append(data[k])
    data=np.array(new_data)
    return data

def cluster_nyc_listings():
    """
    cluster AirBnb listings using k-means++, hierarchical, and GMM
    :return: 
    """
    data=read_dataset("C:/Users/dell/Desktop/listings.csv")
    SSE = []  
    for k in range(1,9):                            #plot the graph to see which k valuse I should choose
        estimator = KMeans(n_clusters=k)            #using elbow method to determine the k value
        estimator.fit(data)
        SSE.append(estimator.inertia_) 
    X = range(1,9)
    plt.xlabel('k')
    plt.ylabel('SSE')
    plt.plot(X,SSE,'o-')
    plt.show()
    # applying elbow method, I choose cluster number to be 3
    
    #kmeans++
    kmeans = KMeans(n_clusters=3, random_state=0).fit(data)
    print(kmeans.cluster_centers_)              #prints the center of the clusters
    print(kmeans.labels_)                       #prints the index of clusters that each data in the sample belongs to
    print(np.unique(kmeans.labels_))            #print the unique index  of clusters
    #hierarichy 
    ward = AgglomerativeClustering(n_clusters=3, linkage='ward').fit(data)
    print(np.unique(ward.labels_))              #print the unique index  of clusters
    print(ward.labels_)                         #prints the index of clusters that each data in the sample belongs to


    #GMM
    gm = GaussianMixture(n_components=3, random_state=0).fit(data)
    print(gm.means_)                            #prints the index of clusters that each data in the sample belongs to
cluster_nyc_listings()
```

### 2b List a few bullet points describing the pros and cons of the various clustering algorithms.

Kmeans++
pros: better initial setup than Kmeans; reduce the influence of outliers; simple, comparatively less running time
cons: needs to manually set the cluster number, the different the cluster number is set, the different the answer; 
Hierarchy
pros:does not need to set the cluster number;
cons:long running time, not good for large sample size; largely influenced by outliers
GMM
pros:it does not require which suportion the data belongs to; fast algorithm, less running time
cons:needs sufficient number of data

### 3b Visualize the clusters by plotting the longitude / lattitude of every listings in a scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram
from sklearn.mixture import GaussianMixture
def visualize_clusters():
    data=read_dataset("C:/Users/dell/Desktop/listings.csv")
    #kmeans
    kmeans= KMeans(n_clusters=3, random_state=0).fit(data) 
    plt.title('Kmeans')
    plt.scatter(data[:,0], data[:,1],c=kmeans.labels_,s=3) 
    plt.show()


    # hierarchy
    model = AgglomerativeClustering(distance_threshold=0, n_clusters=None).fit(data)
    plt.title('Hierarchical Clustering')
    plt.scatter(data[:,0],data[:,1], c=model.labels_,s=3)
    plt.show()

    

    #GMM
    gmm = GaussianMixture(n_components=3).fit(data)
    labels = gmm.predict(data)
    plt.title('GMM')
    plt.scatter(data[:, 0], data[:, 1], c=labels, s=2, cmap='viridis')
    plt.show()

    probs = gmm.predict_proba(data)
    print(probs[:5].round(3))

    size = 50 * probs.max(1) ** 2  # square emphasizes differences
    plt.title('GMM')
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=size)
    plt.show()
visualize_clusters()
```
